```python
import numpy as np
import matplotlib.cm as cm
```

```python
def is_prime(n):
    i = 2
    while i*i<=n:
        if n % i ==0:
            return False
        else:
            i+=1
        
    return True
```

```python
def express_as_binary(i, b):
    result = []
    remaining = i
    for index in range(b+1):
        base = 2**(b-index)
        #print(base)
        if base <= remaining:
            remaining -= base
            result += [1]
        else:
            result += [0]
    return result
```

```python
b = 64
n = 14
print(n)
result = None
for i in range(2**n, 2**(n+1)):
    if is_prime(i):
        print("\t{}\t{}\t{}".format(i, is_prime(i), express_as_binary(i, b)))
        if result is None:
            result=np.asarray(express_as_binary(i, b))
        else:
            result=np.vstack((result, np.asarray(express_as_binary(i, b))))
            
            
print("\t{}\t{}\t{}".format(n, is_prime(n), express_as_binary(n, b)))
```
